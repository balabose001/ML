```python
from sklearn.preprocessing import StandardScaler
st_x =StandardScaler()
```

```python
classifier =DecisionTreeClassifier(criterion='entropy',random_state=0)
```

```python
c_name = ['pregnancies', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
```

```python
df = pd.read_csv('diabetes.csv',header =None,names = c_name)
```

```python
df
```

```python
f_col =['pregnancies', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age']
```

```python
for coln in df:
  df[coln]=df[coln].astype('category').cat.codes
```

```python
df
```

```python
x = df[(f_col)]
```

```python
y =df.label
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3,random_state =1)
```

```python
x_train= st_x.fit_transform(x_train)
```

```python
x_test = st_x.transform(x_test)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```
